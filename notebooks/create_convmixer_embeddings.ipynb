```python
from umap import UMAP
```

```python
data = torch.load(f"{ROOT_DIR_OF_SDMLP_CODEBASE}/ConvMixerEmbeddings/CIFAR10/all_data_test.pt")
umap_2d = UMAP(n_components=2, init='random', random_state=0)

proj_2d = umap_2d.fit_transform(data[0].numpy())

fig_2d = plt.scatter(
    proj_2d[:, 0], proj_2d[:, 1],
    c=data[1].numpy(), cmap="tab10", s=1
)
```
